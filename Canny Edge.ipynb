```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image= cv2.imread(r"F:\Udacity - Computer Vision Nanodegree\CVND_Exercises-master\1_2_Convolutional_Filters_Edge_Detection\images\sunflower.jpg")
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
```

```python
lower = 100
upper = 240
edges = cv2.Canny(gray , lower, upper)
plt.imshow(edges , cmap='gray')
```

```python
width = cv2.Canny(gray, 30,100)
tight= cv2.Canny(gray,180,240)
f, (ax1,ax2) = plt.subplots(1,2, figsize= (20,10))
ax1.set_title("Width")
ax1.imshow(width, cmap='gray')
ax2.set_title("Tight")
ax2.imshow(tight, cmap='gray')
```
